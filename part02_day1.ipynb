## 1.3 이상값 확인

- boston 데이터 세트의 ZN 칼럼을 대상으로 ZN 값의 평균값에서 표준편차의 1.5배보다 크거나 작은 ZN값의 합계를 구하시오.

```python
df = pd.read_csv(data_url)
```

```python
ZN_mean = df.ZN.mean()
```

```python
ZN_std = df.ZN.std()
```

```python
ZN_max = ZN_mean + 1.5 * ZN_std
ZN_min = ZN_mean - 1.5 * ZN_std
```

```python
df['ZN']>ZN_max
```

```python
ZN_new = df.loc[(df['ZN'] > ZN_max) | (df['ZN'] < ZN_min) ]
ZN_new
```

```python
sum(ZN_new['ZN'])
```

## 1.5 순위 구하기
- boston 데이터 세트의 MEDV 칼럼을 기준으로 30번째로 큰 값을 1번 ~ 29번째로 큰 값에 적용한다. 그리고 MEDV 칼럼의 평균값, 중위값, 최소값, 최댓값 순으로 한줄에 출력하시오

```python
df = pd.read_csv(data_url)
```

```python
ans = df.MEDV.sort_values( ascending= False)
```

```python
value = ans.iloc[30]
value
```

```python
ans.iloc[:29] = value
```

```python
ANS = ans.agg(['mean','median', 'min','max'])
```

```python
print(ANS[0],ANS[1],ANS[2],ANS[3])
```

# 2. 복잡한 데이터 분석

## 2.1 그룹별 집계/요약하기
- boston 데이터 세트의 TAX 칼럼이 TAX 칼럼의 중위값보다 큰 데이터를 대상으로, CHAS 칼럼과 RAD 칼럼 순으로 그룹을 지은 후 각 그룹의 데이터 개수를 구하시오. 단, CHAS, RAD 칼럼별 데이터 개수는 COUNT라는 칼럼으로 출력합니다.

```python
data_url = 'https://raw.githubusercontent.com/7ieon/bigData/main/boston.csv'
df = pd.read_csv(data_url)
```

```python
tax_median = df['TAX'].median()
tax_median
```

```python
df.loc[df.TAX>tax_median]
```

```python
df.loc[df.TAX>tax_median][['CHAS','RAD']]
```

```python
new_df = df.loc[df.TAX>tax_median][['CHAS','RAD']]
```

```python
new_df['CHAS'].unique()
```

```python
new_df['RAD'].unique()
```

```python
new2_df = new_df.groupby(['CHAS','RAD'])['RAD'].count()
```

```python
new2_df
```

```python
new3_df = pd.DataFrame(new2_df)
```

```python
new3_df.columns = ['COUNT']
```

```python
new3_df
```
